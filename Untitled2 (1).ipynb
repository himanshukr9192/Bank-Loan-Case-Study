```python
# find missing values in percentage
def missingVals(df):
    null_cols = df.isnull().sum() * 100 / len(df)
    null_cols_df = pd.DataFrame({'percent_missing': null_cols})
    null_cols_df.sort_values(by='percent_missing', ascending=False, inplace=True)
    print(null_cols_df)
```

```python
# dropping columns with more than 40% null values
limitPercen = len(dat) * 0.60
dat = dat.dropna(thresh=limitPercen,axis=1)
```

```python
missingVals(dat)
```

```python
#filling the column occupation_type with not known values because it is not known
dat['OCCUPATION_TYPE'] = dat['OCCUPATION_TYPE'].fillna('NOT_KNOWN')
```

```python
dat.OCCUPATION_TYPE.value_counts()
```

```python
def replaceNull(df,cols):
    for col in cols:
        df[col].fillna(df[col].median(), inplace=True)
```

```python
replaceNull(dat,['EXT_SOURCE_3','AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_MON', 
                 'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_DAY', 'AMT_REQ_CREDIT_BUREAU_HOUR','EXT_SOURCE_2',
                 'AMT_GOODS_PRICE','AMT_ANNUITY','OBS_30_CNT_SOCIAL_CIRCLE', 'DEF_30_CNT_SOCIAL_CIRCLE', 
                 'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE','CNT_FAM_MEMBERS','DAYS_LAST_PHONE_CHANGE'])
dat['NAME_TYPE_SUITE'].fillna(dat.NAME_TYPE_SUITE.mode()[0], inplace=True)
```
